# k-anonymity test

## Import packages

```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr, spearmanr
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

## Census median income data

```python
santa_clara_zip_income_file = "/home/yj/privacy/ACSST5Y2019.S1903_data_with_overlays_2022-02-09T025256.csv"
# santa_clara_zip_income_df = pd.read_csv(santa_clara_zip_income_file, header=[0, 1])
santa_clara_zip_income_df = pd.read_csv(santa_clara_zip_income_file)
santa_clara_zip_income_df.head()
```

```python
santa_clara_zip_income_df["ZCTA"] = santa_clara_zip_income_df['NAME'].str[-5:]
santa_clara_zip_income_sub_df = santa_clara_zip_income_df[
    [
        "ZCTA", 
        "S1903_C03_001E"
    ]
]
santa_clara_zip_income_sub_df.head()
```

## Covid data

```python
santa_clara_zip_covid_file = "/home/yj/privacy/COVID-19_cases_by_zip_code_of_residence_06_05_2022.csv"
santa_clara_covid_df = pd.read_csv(santa_clara_zip_covid_file)
santa_clara_covid_df.head()
```

```python
santa_clara_covid_df["zipcode"] = santa_clara_covid_df["zipcode"].astype(str)
santa_clara_income_covid_df = santa_clara_zip_income_sub_df.merge(
    santa_clara_covid_df,
    how="inner",
    left_on="ZCTA",
    right_on="zipcode"
)
santa_clara_income_covid_df.head()
```

```python
santa_clara_income_covid_df.loc[
    santa_clara_income_covid_df["S1903_C03_001E"] == "250,000+", "S1903_C03_001E"
] = 300000
santa_clara_income_covid_df.loc[
    santa_clara_income_covid_df["S1903_C03_001E"] == "-", "S1903_C03_001E"
] = 0
santa_clara_income_covid_df["S1903_C03_001E"] = santa_clara_income_covid_df["S1903_C03_001E"].astype(float)
santa_clara_income_covid_df.head()
```

```python
# significant, correlated in a reverse direction
correlation_income_rate, p_value_income_rate = pearsonr(santa_clara_income_covid_df["S1903_C03_001E"], santa_clara_income_covid_df["Rate"])
correlation_income_rate, p_value_income_rate
```

```python
santa_clara_income_covid_df.sort_values(by="S1903_C03_001E", ascending=False)
```

## Census demographic data

```python
santa_clara_zip_demo_file = "/home/yj/privacy/ACSDP5Y2019.DP05_data_with_overlays_2022-02-09T122207.csv"
santa_clara_zip_demo_df = pd.read_csv(santa_clara_zip_demo_file)
santa_clara_zip_demo_df.head()
```

```python
santa_clara_zip_demo_df["ZCTA"] = santa_clara_zip_demo_df['NAME'].str[-5:]
santa_clara_zip_demo_sub_df = santa_clara_zip_demo_df[
    [
        "ZCTA", 
        "DP05_0037E", #white
        "DP05_0044E", #asian
        "DP05_0070E", #hispanic
    ]
]
santa_clara_zip_demo_sub_df.head()
```

```python
santa_clara_demo_covid_df = santa_clara_zip_demo_sub_df.merge(
    santa_clara_covid_df,
    how="inner",
    left_on="ZCTA",
    right_on="zipcode"
)
santa_clara_demo_covid_df.head()
```

```python
santa_clara_demo_covid_df["DP05_0037E"]= santa_clara_demo_covid_df["DP05_0037E"].astype(float)
santa_clara_demo_covid_df["DP05_0044E"]= santa_clara_demo_covid_df["DP05_0044E"].astype(float)
santa_clara_demo_covid_df["DP05_0070E"]= santa_clara_demo_covid_df["DP05_0070E"].astype(float)
```

```python
# not significant, not correlated
correlation_demo_white_rate, p_value_demo_white_rate = pearsonr(santa_clara_demo_covid_df["DP05_0037E"], santa_clara_demo_covid_df["Rate"])
correlation_demo_white_rate, p_value_demo_white_rate
```

```python
# not significant, not correlated
correlation_demo_asian_rate, p_value_demo_asian_rate = pearsonr(santa_clara_demo_covid_df["DP05_0044E"], santa_clara_demo_covid_df["Rate"])
correlation_demo_asian_rate, p_value_demo_asian_rate
```

```python
# not significant, not correlated
correlation_demo_hispanic_rate, p_value_demo_hispanic_rate = pearsonr(santa_clara_demo_covid_df["DP05_0070E"], santa_clara_demo_covid_df["Rate"])
correlation_demo_hispanic_rate, p_value_demo_hispanic_rate
```
