# Interval reduction for improving privacy - Covid and Census data

## 1. Import packages and set table display

```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr, spearmanr
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('float_format', '{:.4f}'.format)
```

## 2. Import Census median income data

```python
santa_clara_zip_income_file = "/home/yj/privacy/ACSST5Y2019.S1903_data_with_overlays_2022-02-09T025256.csv"
santa_clara_zip_income_df = pd.read_csv(santa_clara_zip_income_file)
santa_clara_zip_income_df.head()
```

```python
santa_clara_zip_income_df["ZCTA"] = santa_clara_zip_income_df['NAME'].str[-5:]
santa_clara_zip_income_sub_df = santa_clara_zip_income_df[
    [
        "ZCTA", 
        "S1903_C03_001E"
    ]
]
santa_clara_zip_income_sub_df.rename(columns={"S1903_C03_001E": "median_income"}, inplace=True)
santa_clara_zip_income_sub_df.head()
```

## 3. Import Santa Clara Covid data

```python
santa_clara_zip_covid_file = "/home/yj/privacy/COVID-19_cases_by_zip_code_of_residence_06_05_2022.csv"
santa_clara_covid_df = pd.read_csv(santa_clara_zip_covid_file)
santa_clara_covid_df["zipcode"] = santa_clara_covid_df["zipcode"].astype(str)
santa_clara_covid_df.head()
```

```python
santa_clara_covid_df = santa_clara_covid_df[santa_clara_covid_df["Rate"] != 0]
santa_clara_covid_df.sort_values(by="Rate", ascending=False)
```

## 4. Join Census and Covid data

```python
santa_clara_income_covid_df = santa_clara_zip_income_sub_df.merge(
    santa_clara_covid_df,
    how="inner",
    left_on="ZCTA",
    right_on="zipcode"
)
santa_clara_income_covid_df.head()
```

```python
santa_clara_income_covid_df.loc[
    santa_clara_income_covid_df["median_income"] == "250,000+", "median_income"
] = 300000
santa_clara_income_covid_df.loc[
    santa_clara_income_covid_df["median_income"] == "-", "median_income"
] = 0
santa_clara_income_covid_df = santa_clara_income_covid_df[santa_clara_income_covid_df["median_income"] != 0]
santa_clara_income_covid_df["median_income"] = santa_clara_income_covid_df["median_income"].astype(float)
santa_clara_income_covid_df.head()
```

```python
# Joined Covid and Census data distribution
santa_clara_income_covid_df.describe()
```

```python
# Median income and Covid rate - significant, correlated in a reverse direction
correlation_income_rate, p_value_income_rate = pearsonr(santa_clara_income_covid_df["median_income"], santa_clara_income_covid_df["Rate"])
correlation_income_rate, p_value_income_rate
```

```python
# Sort the joined table by the Covid rate in a descending order
santa_clara_income_covid_df.drop_duplicates(["ZCTA", "median_income", "zipcode"], keep="first",inplace=True)
santa_clara_income_covid_df.sort_values(by="Rate", ascending=False)
```

```python
# Correlation heatmap figure
def display_correlation_matrix(data_df):
    # Calculate correlation matrix
    correlation = data_df[["median_income", "Rate"]].corr()
    triu_mask = np.triu(np.ones_like(correlation, dtype=bool))
    
    # matplotlib and seaborn figure and heatmap
    fig, ax = plt.subplots(figsize=(10, 5))
    cmap = sns.diverging_palette(10, 180, as_cmap=True)
    sns.heatmap(correlation, mask=triu_mask, cmap=cmap, vmax=0.5, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .7})

display_correlation_matrix(santa_clara_income_covid_df)
```

```python
# Inverse relation is shown between Census median income and Covid rate
def display_scatter_plot(data_df):
    cmap = sns.color_palette("Set2")
    # Add scatterplot
    sns.scatterplot(
        data=data_df,
        x="median_income", y="Rate",
        palette=cmap, sizes=(10, 200),
    )

display_scatter_plot(santa_clara_income_covid_df)
```

## 5. Additional partitioning - orderly

```python
santa_clara_income_covid_orderly_df = santa_clara_income_covid_df.copy()
conditions = [
    (santa_clara_income_covid_df['median_income'] < 100000),
    ((santa_clara_income_covid_df["median_income"] >= 100000) &
    (santa_clara_income_covid_df["median_income"] < 200000)
    ),
    (santa_clara_income_covid_df['median_income'] >= 200000)
]
values = ["<100000", "100000-200000", "200000+"]
santa_clara_income_covid_orderly_df['median_income']= np.select(conditions, values)
santa_clara_income_covid_orderly_df
```

```python
santa_clara_income_covid_orderly_two_columns_df = santa_clara_income_covid_orderly_df[[
    "median_income",
    "Rate"
]]
santa_clara_income_covid_orderly_two_columns_mean_df = santa_clara_income_covid_orderly_two_columns_df.groupby("median_income").mean()
```

```python
santa_clara_income_covid_orderly_two_columns_mean_df.describe()
```

## 6. Additional partitioning - quantile

```python
santa_clara_income_covid_quantiles_df = santa_clara_income_covid_df.copy()
conditions = [
    (santa_clara_income_covid_df['median_income'] < 102741),
    (
        (santa_clara_income_covid_df["median_income"] >= 102741) &
        (santa_clara_income_covid_df["median_income"] < 127773)
    ),
    (
        (santa_clara_income_covid_df["median_income"] >= 127773) &
        (santa_clara_income_covid_df["median_income"] < 148438)
    ),
    (santa_clara_income_covid_df['median_income'] >= 148438)
]
values = ["<25%", "25-50%", "50-75%", ">75%"]
santa_clara_income_covid_quantiles_df['median_income']= np.select(conditions, values)
santa_clara_income_covid_quantiles_df
```

```python
santa_clara_income_covid_quantiles_two_columns_df = santa_clara_income_covid_quantiles_df[[
    "median_income",
    "Rate"
]]
santa_clara_income_covid_quantiles_two_columns_mean_df = santa_clara_income_covid_quantiles_two_columns_df.groupby("median_income").mean()
```
